```python
with open('data/bundles_all.txt', 'r') as f:
    bundles = f.read().split('\n')
```

```python
bundles[:3]
```

```python
from nltk.tokenize import word_tokenize
```

```python
tokens = [word_tokenize(bundle) for bundle in bundles]
```

```python
tokens[:5]
```

```python
from nltk.stem import WordNetLemmatizer 
```

```python
lemmatizer = WordNetLemmatizer() 
```

```python
lemmatized_bundles = [[lemmatizer.lemmatize(word) for word in token] for token in tokens]
```

```python
lemmatized_bundles[:5]
```

```python
lemmatizer.lemmatize('books')
```

```python
all_lemmatized_bundles = set([' '.join(tokens).lower() for tokens in lemmatized_bundles])
```

```python
all_lemmatized_bundles.remove('')
```

```python
all_lemmatized_bundles
```

```python
print(len(all_lemmatized_bundles), len(bundles))
```

```python
from nltk.stem import PorterStemmer 
```

```python
ps = PorterStemmer() 
```

```python
stemmed_bundles = [[ps.stem(word) for word in token] for token in tokens]
```

```python
stemmed_bundles[:5]
```

```python
all_bundles = set([' '.join(tokens) for tokens in stemmed_bundles])
```

```python
len(all_bundles)
```

```python
all_bundles.remove('')
```

```python
all_bundles
```

```python
bundles
```

```python
with open('data/bundles_all_stemmed.txt', 'w+') as f:
    for bundle in sorted(all_bundles):
        f.write(bundle+'\n')
```

```python
with open('data/bundles_all_lemmatized.txt', 'w+') as f:
    for bundle in sorted(all_lemmatized_bundles):
        f.write(bundle+'\n')
```
